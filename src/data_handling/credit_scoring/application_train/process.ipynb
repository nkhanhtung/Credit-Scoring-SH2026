### 3.2. Xử lý missing value 

```python
"""
Với những cột numeric feature có missing value, ta thay các missing value bằng giá trị median của feature tương ứng
"""
num_impute_cols = ['AMT_ANNUITY', 'CNT_FAM_MEMBERS']

for col in num_impute_cols:
    median_val = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_val)
```

```python
"""
Thay các missing value ở cột NGHỀ NGHIỆP thành Unknown
"""
df_clean['OCCUPATION_TYPE'] = df_clean['OCCUPATION_TYPE'].fillna('Unknown')
df_clean['OCCUPATION_MISSING_FLAG'] = (df_clean['OCCUPATION_TYPE'] == 'Unknown').astype(int)
```

```python
"""
Kiểm tra lại tình trạng missing
"""
print(df_clean.isnull().sum())
```
